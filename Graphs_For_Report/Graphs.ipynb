```python
import numpy as np

import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetika"  # Use a generic serif font
})

x_tag = 0.0
y_tag = 0.0

x = np.linspace(-4, 4, 200)
y = np.linspace(-4, 4, 200)
X, Y = np.meshgrid(x, y)
r = np.sqrt((X - x_tag)**2 + (Y - y_tag)**2)
Z = np.sin(4 * r)  # Example wave pattern

tag_location = np.array([x_tag, y_tag])
end_1 = np.array([-1.9, -2])
start_1 = np.array([-1.2, 0])
end_2 = np.array([-1.2, 2.5])
start_2 = np.array([0.35, 1.06])

plt.figure(figsize=(6, 6))
contour = plt.contourf(X, Y, Z, levels=50,alpha=0.3, cmap='gray')
plt.scatter(x_tag, y_tag, color='black', marker='o', s=50)
plt.scatter(-1.9, -2, color='black', marker='o', s=50)
plt.text(-1.9-1.5,-2-0.5,'$(x_{1}^{end}, y_{1}^{end})$', color='black',fontsize=12)
plt.scatter(-1.2, 0, color='black', marker='o', s=50)
plt.text(-1.2-1.5,0+0.2,'$(x_{1}^{start}, y_{1}^{start})$', color='black',fontsize=12)
plt.plot([-1.9,-1.2],[-2,0], color='black', linestyle='-', linewidth=1.5)
plt.plot([-1.9,x_tag],[-2,y_tag], color='black', linestyle='--', linewidth=1.5)
plt.plot([-1.2,x_tag],[0,y_tag], color='black', linestyle='--', linewidth=1.5)
mid_x = (-1.9 + -1.2) / 2
mid_y = (-2 + 0) / 2
line_length = np.sqrt((-1.9 +1.2)**2 + (-2)**2)
print(line_length)
plt.text(mid_x-0.2, mid_y - 0.5, f'$l_1$: {line_length:.2f}', color='black', fontsize=12, ha='center', rotation=70)
plt.arrow(-1.2, 0, (-1.9 - (-1.2))/1.5,( -2 )/1.5, head_width=0.2, head_length=0.3, fc='black', ec='black', length_includes_head=True)
plt.text(x_tag + 0.2, y_tag , '$(x_{tag}^{2D}, y_{tag}^{2D})$', color='black',fontsize=12)

a = -1.2
b = 2.5
c = 0.35
d = 1.06
plt.scatter(a,b, color='black', marker='o', s=50)
plt.text(a-1,b+0.3,'$(x_{2}^{end}, y_{2}^{end})$', color='black',fontsize=12)
plt.scatter(c, d, color='black', marker='o', s=50)
plt.text(c+0.2,d+0.2,'$(x_{2}^{start}, y_{2}^{start})$', color='black',fontsize=12)
plt.plot([a,c],[b,d], color='black', linestyle='-', linewidth=1.5)
plt.plot([a,x_tag],[b,y_tag], color='black', linestyle='--', linewidth=1.5)
plt.plot([c,x_tag],[d,y_tag], color='black', linestyle='--', linewidth=1.5)
mid_x = (a+c) / 2
mid_y = (b + 1.2) / 2
line_length = np.sqrt((a-c)**2 + (b-d)**2)
print(line_length)
plt.text(mid_x+0.2, mid_y-0.2, f'$l_2$: {line_length:.2f}', color='black', fontsize=12, ha='center', rotation=-45)
plt.arrow(c, d, (a - c)/1.5,( b - d)/1.5, head_width=0.2, head_length=0.3, fc='black', ec='black', length_includes_head=True)
plt.xlim = (-5, 5)
plt.ylim = (-5, 5)

plt.title('Different paths example', fontsize=16)

#plt.axis('off')  # Turn off axis numbering and labels
plt.show()


plt.figure(figsize=(6, 6))

dy = end_2[1] - start_2[1]
dx = end_2[0] - start_2[0]

theta = np.arctan2(dy, dx)
rotation_matrix = np.array([[np.cos(theta), np.sin(theta)], [-np.sin(theta), np.cos(theta)]])

tag_loc_new = np.dot(rotation_matrix, tag_location - start_2)
start_2_new = np.dot(rotation_matrix, start_2 - start_2)
end_2_new = np.dot(rotation_matrix, end_2 - start_2)
print(end_2_new)

x = np.linspace(-1, 4, 200)
y = np.linspace(-1, 4, 200)
X, Y = np.meshgrid(x, y)
r = np.sqrt((X - tag_loc_new[0])**2 + (Y - tag_loc_new[1])**2)
Z = np.sin(4 * r)  # Example wave pattern
contour = plt.contourf(X, Y, Z, levels=50,alpha=0.3, cmap='gray')
plt.scatter(tag_loc_new[0], tag_loc_new[1], color='black', marker='o', s=50)
plt.scatter(start_2_new[0], start_2_new[1], color='black', marker='o', s=50)
plt.text(start_2_new[0]-0.5,start_2_new[1]-0.2,'$(\hat{x}^{start}, \hat{y}^{start})$', color='black',fontsize=12)
plt.scatter(end_2_new[0], end_2_new[1], color='black', marker='o', s=50)
plt.plot([start_2_new[0], end_2_new[0]],[start_2_new[1], end_2_new[1]], color='black', linestyle='-', linewidth=1.5)
plt.arrow(start_2_new[0], start_2_new[1], (end_2_new[0] - start_2_new[0])/1.5,( end_2_new[1] - start_2_new[1])/1.5, head_width=0.1, head_length=0.15, fc='black', ec='black', length_includes_head=True)
plt.plot([start_2_new[0], tag_loc_new[0]],[start_2_new[1], tag_loc_new[1]], color='black', linestyle='--', linewidth=1.5)
plt.plot([end_2_new[0], tag_loc_new[0]],[end_2_new[1], tag_loc_new[1]], color='black', linestyle='--', linewidth=1.5)
plt.text(end_2_new[0]-0.5,end_2_new[1]-0.2,'$(\hat{x}^{end}, \hat{y}^{end})$', color='black',fontsize=12)
plt.text(tag_loc_new[0] + 0.2, tag_loc_new[1] , '$(\hat{x}_{tag}^{2D}, \hat{y}_{tag}^{2D})$', color='black',fontsize=12)
plt.title('Transformed paths example', fontsize=16)
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Create a 3D figure
fig = go.Figure()

# Generate a random 2D path for the antenna with a fixed positive z value
antenna_path_x = np.linspace(start_2_new[0], end_2_new[0], 100)  # X-coordinates for the path
antenna_path_y = np.linspace(start_2_new[1], end_2_new[1], 100)  # Y-coordinates for the path
antenna_path_z = np.full_like(antenna_path_x, 0)  # Fixed positive Z-coordinate

# Generate a random RFID tag location in 3D space with a positive z value
rfid_tag_x = tag_loc_new[0]
rfid_tag_y = tag_loc_new[1]
rfid_tag_z = 1.5



# Add a solid line from start point to end point
fig.add_trace(go.Scatter3d(
    x=[start_2_new[0], end_2_new[0]],
    y=[start_2_new[1], end_2_new[1]],
    z=[0, 0],
    mode='lines',
    line=dict(color='black', width=2),
    name='Antenna Path'
))

# Add the extended path in the negative x direction with dashed lines
fig.add_trace(go.Scatter3d(
    x=np.linspace(start_2_new[0], start_2_new[0] - 1, 50),  # Extended X-coordinates
    y=np.full(50, start_2_new[1]),  # Keep Y-coordinate constant
    z=np.full(50, 0),  # Keep Z-coordinate constant
    mode='lines',
    line=dict(color='darkgray', width=2, dash='dot'),
    name='Extended Antenna Path'
))
# Add annotations for the 3D and 2D RFID tag locations
fig.add_trace(go.Scatter3d(
    x=[rfid_tag_x],
    y=[rfid_tag_y],
    z=[rfid_tag_z],
    mode='text',
    text=["RFID Tag (3D)"],
    textposition="top center",
    name='Annotation'
))

fig.add_trace(go.Scatter3d(
    x=[rfid_tag_x_2d],
    y=[np.sqrt(rfid_tag_y**2 + rfid_tag_z**2)],
    z=[0],
    mode='text',
    text=["RFID Tag (2D)"],
    textposition="top center",
    name='Annotation'
))
fig.add_trace(go.Scatter3d(
    x=[rfid_tag_x],
    y=[rfid_tag_y],
    z=[rfid_tag_z],
    mode='markers',
    marker=dict(size=5, color='black'),
    name='RFID Tag (3D)'
))

# Add points for the RFID tag (2D)
fig.add_trace(go.Scatter3d(
    x=[rfid_tag_x_2d],
    y=[rfid_tag_y_2d],
    z=[rfid_tag_z_2d],
    mode='markers',
    marker=dict(size=5, color='black'),
    name='RFID Tag (2D)'
))

# Add points for the start of the antenna path
fig.add_trace(go.Scatter3d(
    x=[start_2_new[0]],
    y=[start_2_new[1]],
    z=[0],
    mode='markers',
    marker=dict(size=5, color='black'),
    name='Start Point'
))

# Add points for the end of the antenna path
fig.add_trace(go.Scatter3d(
    x=[end_2_new[0]],
    y=[end_2_new[1]],
    z=[0],
    mode='markers',
    marker=dict(size=5, color='black'),
    name='End Point'
))
# Add annotations for the start and end points of the antenna path
fig.add_trace(go.Scatter3d(
    x=[start_2_new[0]],
    y=[start_2_new[1]],
    z=[0],
    mode='text',
    text=["Start Point"],
    textposition="top center",
    name='Annotation'
))

fig.add_trace(go.Scatter3d(
    x=[end_2_new[0]],
    y=[end_2_new[1]],
    z=[0],
    mode='text',
    text=["End Point"],
    textposition="top center",
    name='Annotation'
))

# Add point with same x as 2D tag location on the extended path
fig.add_trace(go.Scatter3d(
    x=[rfid_tag_x_2d],
    y=[0],
    z=[0],
    mode='markers',
    marker=dict(size=5, color='lightgray'),
    name='RFID Tag (Same X)'
))

# Add trace between 3D tag and 2D tag location
fig.add_trace(go.Scatter3d(
    x=[rfid_tag_x, rfid_tag_x_2d],
    y=[rfid_tag_y, rfid_tag_y_2d],
    z=[rfid_tag_z, rfid_tag_z_2d],
    mode='lines',
    line=dict(color='darkgray', width=2),
    name='Tag Connection'
))

# Add trace between 2D tag and antenna path
fig.add_trace(go.Scatter3d(
    x=[rfid_tag_x_2d, rfid_tag_x_2d],
    y=[rfid_tag_y_2d, 0],
    z=[rfid_tag_z_2d, 0],
    mode='lines',
    line=dict(color='darkgray', width=2),
    name='Tag to Antenna Path'
))

# Add trace between 3D tag and antenna path
fig.add_trace(go.Scatter3d(
    x=[rfid_tag_x, rfid_tag_x_2d],
    y=[rfid_tag_y, 0],
    z=[rfid_tag_z, 0],
    mode='lines',
    line=dict(color='darkgray', width=2),
    name='Tag to Antenna Path'
))

# Add point with same x but the sqrt of y and z of 3D tag location
fig.add_trace(go.Scatter3d(
    x=[rfid_tag_x],
    y=[np.sqrt(rfid_tag_y**2 + rfid_tag_z**2)],
    z=[0],
    mode='markers',
    marker=dict(size=5, color='gray'),
    name='RFID Tag (2D converted))'
))

# Add circular trace that starts from the 3D tag location and follows a circle
phi = np.arccos(rfid_tag_z / np.sqrt(rfid_tag_y**2 + rfid_tag_z**2))
theta = np.linspace(0, np.pi/2 - phi, 100)  # Angle values for the circle
radius = np.sqrt(rfid_tag_y**2 + rfid_tag_z**2)  # Radius of the circle
circle_x = np.full_like(theta, rfid_tag_x)  # X-coordinate remains constant
circle_y = radius * np.cos(theta)  # Y-coordinates of the circle
circle_z = radius * np.sin(theta)  # Z-coordinates of the circle

# Add the circular trace to the plot
fig.add_trace(go.Scatter3d(
    x=circle_x,
    y=circle_y,
    z=circle_z,
    mode='lines',
    line=dict(color='gray', width=2),
    name='Circular Path'
))

# Extend the antenna path in the negative x direction with dotted lines
extended_path_x = np.linspace(start_2_new[0] - 1, start_2_new[0], 50)  # Extended X-coordinates
extended_path_y = np.full_like(extended_path_x, start_2_new[1])  # Keep Y-coordinate constant
extended_path_z = np.full_like(extended_path_x, 0)  # Keep Z-coordinate constant

# Add the extended path to the plot
fig.add_trace(go.Scatter3d(
    x=extended_path_x,
    y=extended_path_y,
    z=extended_path_z,
    mode='lines',
    line=dict(color='lightgray', width=2, dash='dot'),
    name='Extended Antenna Path'
))

# Add text annotations instead of the legend
annotations = [
    dict(x=rfid_tag_x, y=rfid_tag_y, text="RFID Tag (3D)", showarrow=False, font=dict(color="black")),
    dict(x=rfid_tag_x_2d, y=rfid_tag_y_2d, text="RFID Tag (2D)", showarrow=False, font=dict(color="black")),
    dict(x=start_2_new[0], y=start_2_new[1], text="Start Point", showarrow=False, font=dict(color="black")),
    dict(x=end_2_new[0], y=end_2_new[1], text="End Point", showarrow=False, font=dict(color="black"))
]

# Update the layout to reflect the changes
fig.update_layout(
    title='Antenna 2D Path and RFID Tag in 3D Space',
    scene=dict(
        xaxis=dict(title='X'),
        yaxis=dict(title='Y'),
        zaxis=dict(title='Z')
    ),
    showlegend=False,  # Remove the legend
    annotations=annotations
)

# Show the updated figure
fig.show(renderer="browser")
```
